```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm

data_svm = pd.read_excel("Data_SVM.xlsx")
data_svm["X1"] = data_svm["X1"].astype(float)
data_svm["X2"] = data_svm["X2"].astype(float)
data_svm["Y"] = data_svm["Y"].astype(int)
```

```python
scatter = plt.scatter(data_svm['X1'],data_svm["X2"] , c=data_svm['Y'])
classes = ['Class 1', 'Class 2']
plt.legend(handles=scatter.legend_elements()[0], labels=classes)
plt.show()
```

```python
model = svm.SVC(kernel='linear', C=1.0)
clf = model.fit(data_svm[['X1','X2']], data_svm['Y'])
w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(0, 20)
yy = a * xx - (clf.intercept_[0]) / w[1]
plt.plot(xx, yy, 'k-')
scatter1 = plt.scatter(data_svm["X1"], data_svm["X2"], c=data_svm['Y'])
classes = ['Class 1', 'Class 2']
plt.legend(handles=scatter1.legend_elements()[0], labels=classes)
w_hat = clf.coef_[0] / (np.sqrt(np.sum(clf.coef_[0] ** 2)))
margin = 1 / np.sqrt(np.sum(clf.coef_[0] ** 2))
decision_boundary_points = np.array(list(zip(xx, yy)))
points_of_line_above = decision_boundary_points + w_hat * margin
points_of_line_below = decision_boundary_points - w_hat * margin
plt.plot(points_of_line_above[:, 0], points_of_line_above[:, 1], 'b--', linewidth=2)
plt.plot(points_of_line_below[:, 0], points_of_line_below[:, 1], 'g--', linewidth=2)
plt.show()
```

```python
print("Prediction for (8,15) is: ", clf.predict([[8,15]]))
print("Prediction for (7,4) is: ", clf.predict([[7,4]]))
```
